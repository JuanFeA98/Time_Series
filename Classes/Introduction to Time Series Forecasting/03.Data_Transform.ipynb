## **Librer√≠as**

```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = False
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import boxcox
```

## **Datos**

```python
airpasengers = pd.read_csv(
    '../../Data/AirPassengers.csv',
    header=0,
    index_col=0,
    parse_dates=True
)

airpasengers.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(14, 6))

fig.patch.set_facecolor('#21252b')

ax[0].plot(airpasengers)
ax[0].set_title('N√∫mero de pasajeros a lo largo del tiempo', color='white', weight=700)
ax[0].set_facecolor('#282c34')
ax[0].tick_params(colors='white')
ax[0].set_xlabel('A√±o', color='white')
ax[0].set_ylabel('Cantidad de vuelos', color='white')

ax[1].hist(airpasengers)
ax[1].set_title('Cantidad de vuelos por n√∫mero de pasajeros', color='white', weight=700)
ax[1].set_facecolor('#282c34')
ax[1].tick_params(colors='white')
ax[1].set_xlabel('N√∫mero de pasajeros', color='white')
ax[1].set_ylabel('Cantidad de vuelos', color='white')

plt.subplots_adjust(hspace=0.7)

plt.show()
```
